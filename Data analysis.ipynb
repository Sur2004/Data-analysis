```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\Gamma\Downloads\Students Social Media Addiction.csv")


def age_group(age):
    if age < 18:
        return 'Under 18'
    elif 18 <= age <= 24:
        return '18-24'
    elif 25 <= age <= 34:
        return '25-34'
    elif 35 <= age <= 44:
        return '35-44'
    elif 45 <= age <= 54:
        return '45-54'
    else:
        return '55+'

df['Age_Group'] = df['Age'].apply(age_group)
sns.set(style="whitegrid")

plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Age_Group', order=df['Age_Group'].value_counts().index)
plt.title("Age Group Distribution of Students")
plt.xlabel("Age Group")
plt.ylabel("Number of Students")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Age_Group', y='Avg_Daily_Usage_Hours', palette='coolwarm')
plt.title("Daily Social Media Usage by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Avg. Daily Usage (hours)")
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Most_Used_Platform', hue='Age_Group', palette='viridis')
plt.title("Most Used Platforms by Age Group")
plt.xlabel("Platform")
plt.ylabel("Number of Students")
plt.legend(title='Age Group')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Age_Group', y='Addicted_Score', palette='Set3')
plt.title("Addiction Score by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Addiction Score (0-10)")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Addicted_Score', y='Mental_Health_Score', hue='Age_Group', palette='magma')
plt.title("Mental Health Score vs Addiction Score")
plt.xlabel("Addiction Score")
plt.ylabel("Mental Health Score (0-10)")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Avg_Daily_Usage_Hours', y='Sleep_Hours_Per_Night', hue='Age_Group', palette='cubehelix')
plt.title("Sleep Hours vs Daily Social Media Usage")
plt.xlabel("Daily Usage Hours")
plt.ylabel("Sleep Hours Per Night")
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Age_Group', y='Conflicts_Over_Social_Media', palette='Pastel1')
plt.title("Conflicts Over Social Media by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Number of Conflicts")
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Gender', y='Avg_Daily_Usage_Hours', hue='Age_Group', palette='Accent')
plt.title("Gender-wise Daily Social Media Usage")
plt.xlabel("Gender")
plt.ylabel("Avg. Daily Usage (hours)")
plt.legend(title="Age Group")
plt.tight_layout()
plt.show()
```
